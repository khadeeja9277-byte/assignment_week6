```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
marks= np.random.randint(30,101, size=(100,5))
columns= ["student_id","maths","science","english","history","computer"]
df= pd.DataFrame(marks,columns=columns[1:])
df["student_id"] = range(1,101)
df["average"]= df[columns].mean(axis=1)
def get_grade(x):
    if x >=85:
        return "A"
    elif x>=70:
        return "B"
    elif x>=50:
        return "C"
    else:
        return"D"

df["grade"] =df["average"].apply(get_grade)
# Distribution of grades bar chart
sns.countplot(data=df, x= "grade")
plt.title('distribution of grades')
plt.xlabel('grades')
plt.ylabel('number of students')
plt.show()
```
